```python
from datasets import load_dataset

# Load and select the first 100 examples
full_dataset = load_dataset("tasksource/Boardgame-QA")
```

```python
subset_dataset = full_dataset['train'].select(range(5))
# you can edit and put in your own range
```

```python
import getpass

# https://platform.openai.com/account/api-keys
secret_key = getpass.getpass('Please enter your openai key: ')
```

```python
from openai import OpenAI
client = OpenAI(api_key=secret_key)

def get_response(prompt):
  response = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[{
      "role": "user",
      "content": f"{prompt}"
    }]
    )
  return response.choices[0].message.content
```

```python
with open('subquestions.txt','w') as of:
  for i in range(0,3):
    scenario= subset_dataset[i]['example']
    lhs='Given a scenario and its rules: '
    rhs=' Generate subquestions which can help answer the final question in the last line of the scenario following the format: subquestion1) subquestion2)'
    prompt=lhs + scenario + rhs
    subquestions=get_response(prompt)
    of.write("{}".format(i+1))
    of.write('\n')
    of.write(subquestions)
    of.write('\n')
```
